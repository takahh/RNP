```python
import pandas as pd
from matplotlib import pyplot as plt
path = '/Users/tkimura/Desktop/RNP/check_contact/hbond_summary.txt'
# path = '/Users/tkimura/Desktop/RNP/check_contact/hbond_summary (original).txt'
figpath = '/Users/tkimura/Desktop/RNP/check_contact/'
title = 'All Hbonding Chains Except the Two'
title3 = 'All Hbonding Chains Except the Two \n(more than 2 hbonds)'

# plot a pi chart to show how many other chains are related to hbonds

df = pd.read_csv(path, sep='\t')
label_num = len(df['allchains'].value_counts())
label_num3 = len(df['allchains3'].value_counts())

labellist, labellist3 = [], []
for pair in ([labellist, label_num, 'allchains'], [labellist3, label_num3, 'allchains3']):
    series = df[pair[2]].value_counts().sort_index()
    for i in range(0, pair[1]):
        if series.values[i] > 5000:
            pair[0].append(series.index[i])
        else:
            pair[0].append('')

fig = plt.figure()
plt.pie(df['allchains'].value_counts().sort_index(),labels=labellist, textprops={'fontsize': 12})
plt.title(f'{title}')
fig.savefig(f'{figpath}all_touchingchains.jpg', dpi=500, bbox_inches="tight")
print(df['allchains'].count())
plt.show()

plt.pie(df['allchains3'].value_counts().sort_index(),labels=labellist3, textprops={'fontsize': 12})
plt.title(f'{title3}')
print(df['allchains'].count())
fig.savefig(f'{figpath}morethan2_touchingchains.jpg', dpi=500, bbox_inches="tight")
plt.show()
plt.close(fig)
plt.close()
```
